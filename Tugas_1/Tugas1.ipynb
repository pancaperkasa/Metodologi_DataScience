### No 3
#### Makanan = Bar Chart

Alasan Menggunakan Pie Chart Dikarenakan studi kasus no 3 hanya ingin mengetahui makanan penutup yang kurang peminatnya, sehingga saya menggunakan visualisasi Bar Chart karena bisa menampilkan bar yang bisa dilakukan sorting mulai terendah sampai tertinggi.

Sebutkan tiga makanan penutup yang harus disingkirkan ?

Berdasarkan hasil pemodelan dibawah, maka didapatkan grafik visualiasi yang menunjukkan 3 makanan penutup yang kurang diminati adalah Kue Wajik, Pastel, dan Puding Vanilla.

```python
makananpenutup = [
    'Donat', 'Pastel', 'Kue Coklat', 
    'Ice Cream', 'Puding Vanilla', 'Brownies', 
    'Puding Strawberry', 'Puding Coklat', 'Ice Cream Nutella', 
    'Kue Coklat Keju', 'Kue Wajik', 'Kue Sus', 'Mochi' 
]

datapenjualan = [14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9,13]

df = pd.DataFrame({'Dessert': makananpenutup, 'Penjualan': datapenjualan,})
df.sort_values(by= 'Penjualan', inplace = True)

kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna [0] = '#FF0000'
warna [1] = '#FF0000'
warna [2] = '#FF0000'

plt.figure(figsize=(25,10))
plt.bar(kor_x, df['Penjualan'], tick_label=df['Dessert'], color=warna)
plt.xticks(rotation=90)
plt.ylabel("Penjualan")
plt.title('3 Dessert Paling Kurang Di Minati')
plt.show()
```
