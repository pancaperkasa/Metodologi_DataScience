### Imputasi

Imputasi adalah mengganti nilai/data yang hilang (missing value; NaN; blank) dengan nilai pengganti

### Mean
Nilai rata - rata

```python
import pandas as pd
import numpy as np

kolom = {'col1':[2, 9, 19], 
         'col2':[5, np.nan, 17],
         'col3':[3, 9, np.nan],
         'col4':[6, 0, 9],
         'col5':[np.nan, 7, np.nan]
        }

data = pd.DataFrame(kolom)
```

```python
data
```

```python
data.fillna(data.mean())
```

### Arbitary
Nilai suka - suka

```python
umur = {'umur': [29, 43, np.nan, 25, 34, np.nan, 50]}

data = pd.DataFrame(umur)
data
```

```python
data.fillna(99)
```

### End Of Tail

```python
umur = {'umur': [29, 43, np.nan, 25, 34, np.nan, 50]}

data = pd.DataFrame(umur)
data
```

Bagaimana dengan data kategorikal ?
Kita akan gunakan imputasi modus

### Modus

```python
from sklearn.impute import SimpleImputer

mobil = {'merk':['Ford', 'Ford', 'Toyota', 'Honda', np.nan, 'Toyota', 'Honda', 'Toyota', np.nan, np.nan]}

data = pd.DataFrame(mobil)
```

```python
data
```

```python
imp = SimpleImputer(strategy= 'most_frequent')
```

```python
imp.fit_transform(data)
```

Bagaimana kalau kita butuh untuk mengatasi missing value untuk kategorikal dan numerikal secara bersamaan ?
kita bisa gunakan random sample

### Random Sample

```python
#import dulu si Random Sample
from feature_engine.imputation import RandomSampleImputer

#silahkan buat data yang ada missing value-nya
data = {
    'Jenis Kelamin': ['laki', 'perempuan', 'laki', 'laki', np.nan, np.nan, 'perempuan', np.nan, 'perempuan','laki'],
    'Umur': [22, np.nan, np.nan, 25, 44, 38, 28,53, np.nan,34]
}

dataPenduduk = pd.DataFrame(data)
dataPenduduk
```

```python
#Buat Imputer
impPenduduk = RandomSampleImputer(random_state=29)

#fit 
impPenduduk.fit(dataPenduduk)

#Ubah Data
hasilPenduduk = impPenduduk.transform(dataPenduduk)

#Tampil Data
hasilPenduduk
```
